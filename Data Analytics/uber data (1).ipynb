```python
uber_apr=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-apr14.csv')
uber_may=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-may14.csv')
uber_jun=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-jun14.csv')
uber_jul=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-jul14.csv')
uber_aug=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-aug14.csv')
uber_sep=pd.read_csv('H:/MY COMPUTER/Personal documents/cadc AI/Notes/Data Analytics/uber-raw-data-sep14.csv')
```

```python
print("Total number of booking on April=",len(uber_apr))
print("Total number of booking on May=",len(uber_may))
print("Total number of booking on June=",len(uber_jun))
print("Total number of booking on July=",len(uber_jul))
print("Total number of booking on August=",len(uber_aug))
print("Total number of booking on September=",len(uber_sep))
```

```python
frames=[uber_apr,uber_may,uber_jun,uber_jul,uber_aug,uber_sep]
total_data=pd.concat(frames)
total_data.shape
```

```python
print("Total number of booking from April to September=",len(total_data))
```

```python
total_data['Base'].value_counts()
```

```python
total_data['Date/Time']=total_data['Date/Time'].apply(pd.to_datetime)
total_data['date']=total_data['Date/Time'].apply(lambda total_data:total_data.day)
total_data['hour']=total_data['Date/Time'].apply(lambda total_data:total_data.hour)
total_data['e']=1
```

```python
total_data['month']=total_data['Date/Time'].apply(lambda total_data:total_data.month)
```

```python
print(total_data.groupby(['hour','Base']).count()['e'])
```

```python
month=total_data['month'].unique()
daymonth=total_data.groupby(['month','date']).count()['e'].unstack(level=1)
daymonth.index=pd.Categorical(daymonth.index,categories=month,ordered=True)
daymonth.index.name='monthofdate'

daymonth.sort_index(inplace=True)
daymonth
```
