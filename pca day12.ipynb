```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as r
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
var=['var'+str(i) for i in range(0,101)]
print(var)
```

```python
s1=['striker' + str(i) for i in range(1,6)]
s2=['defender' + str(i) for i in range(1,6)]
```

```python
s1
```

```python
s2
```

```python
df=pd.DataFrame(columns=[*s1,*s2],index=var)
```

```python
df
```

```python
for var1 in df.index:
    df.loc[var1,'striker1':'striker5']=np.random.poisson(lam=r.randrange(10,1000),size=5)
    df.loc[var1,'defender1':'defender5']=np.random.poisson(lam=r.randrange(10,1000),size=5)
```

```python
df
```

```python
sd=preprocessing.scale(df.T)
```

```python
sd
```

```python
pc=PCA()
pc.fit(sd)
```

```python
pcd=pc.transform(sd)
pcd
```

```python
pr=np.round(pc.explained_variance_ratio_*100,decimals=1)
```

```python
pr
```

```python
label=['pc'+str(i) for i in range(1,len(pr)+1)]
label
```

```python
plt.bar(x=range(1,len(pr)+1),height=pr,label=label)
plt.xlabel('Principal component')
plt.ylabel('% of experiment')
plt.show()
```

```python
pca_df=pd.DataFrame(pcd,index=[*s1,*s2],columns=label)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
```

```python
features=sd.fit_transform(df)
```

```python
pcd=pc.transform(features)
pcd
```

```python
pca = PCA()

principalComponents = pca.fit_transform(pcd)
principalComponents
```

```python
pr=np.round(pca.explained_variance_ratio_*100,decimals=1)
pr
```

```python
label=['pc'+str(i) for i in range(1,len(pr)+1)]
plt.bar(x=range(1,len(pr)+1),height=pr,label=label)
plt.xlabel('Principal component')
plt.ylabel('% of experiment')
plt.show()
```

```python
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(features)
principalComponents

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
outcome=pd.DataFrame(data.target,columns=['outcome'])
finalDf = pd.concat([principalDf, outcome], axis = 1)
finalDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)


targets = [1,0]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['outcome'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
